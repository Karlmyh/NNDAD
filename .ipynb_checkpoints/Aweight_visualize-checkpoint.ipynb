```python
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath_seq = glob.glob("./results/anomaly_realdata/*.npy")
filepath_seq.sort()
```

```python
for filepath in filepath_seq:
    weight = np.load(filepath)
    name = filepath.split("_")[1][:-4]
    plt.figure(figsize=(4, 4), dpi=80)
    plt.title(name)
    plt.plot(weight)
    plt.show()
```
